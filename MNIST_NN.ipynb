## Part A: Train an XOR network using PyTorch

**Goal:** Use PyTorch to train a neural network that correctly predicts XOR for all 4 inputs (**4 out of 4 correct**).

**Requirements**
- Use a **Binary Cross Entropy** loss (e.g., `nn.BCELoss()` or `nn.BCEWithLogitsLoss()`).
- You must **train** the network (do not manually set weights).
- Show the final predictions and the final accuracy.


```python
# Imports
import torch
import torch.nn as nn
```

### A1) Create the XOR dataset

Fill in the XOR input points and labels.


```python
# TODO: define the XOR inputs (shape: 4 x 2) as float tensors
X = torch.tensor([
     [0., 0.],
    [1., 0.],
    [0., 1.],
    [1., 1.],
])

# TODO: define the XOR labels (shape: 4 x 1) as float tensors
y_true = torch.tensor([
    [0.],
    [1.],
    [1.],
    [0.],
])

X
```

### A2) Define your model

Hints:
- XOR is not linearly separable, so you need **at least one hidden layer** and a **nonlinearity** (ReLU, Tanh, etc.).
- If you use `nn.BCELoss()`, the model output should be in (0,1), so include a `Sigmoid()` at the end.
- If you use `nn.BCEWithLogitsLoss()`, **do not** use `Sigmoid()` in the model.


```python
# TODO: build a small network for XOR
# Example structure (edit it): Linear -> activation -> Linear -> (Sigmoid)

#1) d_in = [x1,x2], 2 features (per entry)
d_in = 2

#2) Needed 8 nodes (neurons) in first hidden layer.
d_hidden = 8

#3) 1 output (probability of class 1/0) per entry.
d_out = 1

model = nn.Sequential(
    # Step-1: Pre-Activation Layer. Y = W*X^T + b
    nn.Linear(d_in, d_hidden),
    # Step-2: Apply ReLU activation function (vectorized) : Y1=ReLU(Y,*)
    nn.ReLU(),
    # Step-3: Second Affine Map: W2 of shape (d_hidden X d_out).
    nn.Linear(d_hidden, d_out),
    # Output: Y2 = Y1*W2 + b, where Y2 is (batch_size x d_out), a 1-D tensor of batch_size scalar entries.
    # Each of these entries is a probability, using BCEWithLogitsLoss
    )

model

```

### A3) Train the model

Requirements:
- Choose an optimizer (SGD or Adam).
- Train long enough to get 4/4 correct.
- Print (or store) the loss occasionally so we can see training progress.

Tip: If SGD is unstable, try Adam, or change the learning rate.


```python
# Choose a loss function
loss_fn = nn.BCEWithLogitsLoss()

# Choose an optimizer: Adam is often more robust than SGD for getting unstuck from local minima.
# A common starting learning rate for Adam is 0.001 or 0.01.
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# Define the number of training steps (epochs or iterations)
num_steps = 10000

# Training loop
for step in range(num_steps):
    # 1. Forward pass: Compute predicted y by passing X to the model
    # This calculates the output of the neural network for the given inputs.
    y_hat = model(X)

    # 2. Compute loss: Calculate the difference between predictions (y_hat) and true labels (y_true)
    # The BCEWithLogitsLoss function handles the sigmoid and BCELoss together for stability.
    loss = loss_fn(y_hat, y_true)

    # 3. Backpropagation: Calculate gradients
         #i) Zero gradients: Clear previous gradients before computing new ones
         #Needed since PyTorch accumulates gradients by default.
    optimizer.zero_grad()
        #ii) Backward pass: Compute gradients of the loss with respect to all learnable parameters
        #This populates the .grad attribute for each parameter in the model.
    loss.backward()
       #iii) Optimizer step: Update the model's parameters using the calculated gradients

    # 4. The Adam optimizer adjusts weights based on adaptive learning rates.
    optimizer.step()

    # Print progress every 500 steps to monitor loss.
    if step % 500 == 0:
        print(f"Step {step}, Loss: {loss.item():.4f}")

print(f"Final Loss after {num_steps} steps: {loss.item():.4f}")
```

### A4) Evaluate XOR accuracy

Compute:
- predicted probabilities
- predicted labels (use threshold 0.5)
- accuracy out of 4


```python
# compute probabilities (or logits) and convert to predicted labels
with torch.no_grad():
    y_hat = model(X) # Now y_hat contains logits, not probabilities

# if y_hat are logits, apply sigmoid first, then threshold at 0.5
probabilities = torch.sigmoid(y_hat) # Apply sigmoid to convert logits to probabilities
y_pred = (probabilities >= 0.5).float() #P(class=1|input)

# compute accuracy
accuracy = (y_pred == y_true).float().mean()

print("y_hat (logits) =", y_hat)
print("probabilities =", probabilities)
print("y_pred =", y_pred)
print("y_true =", y_true)
print("Accuracy =", accuracy)
```

Ans:
1. The architecture: (Linear) $\rightarrow$ (ReLU) $\rightarrow$ (Linear). I used just one linear hidden layer ("pre-activation"), with 8 nodes (the corresponding weight matrix has shape $8\times2$).
- I started out with 4, but realized that I needed more nodes for increased accuracy (which stalled at 75% with only 4 layers and SGD).

## Part B: Train an MNIST network using PyTorch

**Goal:** Adapt the PyTorch tutorial approach to train an MNIST classifier and reach **at least 91% test accuracy**.

Use `datasets.MNIST` (not FashionMNIST).

Recommended reference:
- PyTorch quickstart tutorial (structure for loaders/train/test loops).

**What to turn in**
- Your code.
- Final test accuracy (as a number).
- A short answer (4–8 sentences) describing what you changed to reach ≥91%.


```python
from torchvision import datasets
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader
```

### B1) Load MNIST

Note: MNIST is already split into train (60,000 images) and test sets (10,000 images). We will load the training_data and test_data here.

```python
training_data = datasets.MNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor()
)

test_data = datasets.MNIST(
    root="data",
    train=False,
    download=True,
    transform=ToTensor()
)
```

### B2) Create DataLoaders

Choose a batch size and create:
- `train_dataloader`
- `test_dataloader`


```python
# batch_size = number of images used to update parameters for each iteration of gradient descent.
# Calculate per-sample average loss and gradient for a sample of size 64.
batch_size = 64

train_dataloader = DataLoader(training_data, batch_size=batch_size)
test_dataloader = DataLoader(test_data, batch_size=batch_size)

# inspect one batch
for X, y in test_dataloader:
    print(X.shape, y.shape)
    break
```

### B3) Define your model

Hints:
- A simple start is: `Flatten -> Linear -> ReLU -> Linear -> ReLU -> Linear(10)`.
- You can also try other architectures, but keep it reasonably small.
- For MNIST classification with `nn.CrossEntropyLoss()`, the model should output **raw scores** (logits) of shape `(batch, 10)` and you should **not** apply softmax in the model.


```python
# Define model
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.layers = nn.Sequential(
            nn.Linear(784, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.layers(x)
        return logits

model = NeuralNetwork()
print(model)
```

### B4) Loss + optimizer

Use `nn.CrossEntropyLoss()`.

Pick an optimizer (SGD or Adam) and a learning rate.


```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

### B5) Training and testing loops

Fill in `train()` and `test()`.

Hints:
- `model.train()` during training, `model.eval()` during testing.
- Wrap testing in `torch.no_grad()`.
- Accuracy: compare `pred.argmax(1)` to labels `y`.


```python
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    model.train()
    for batch, (X, y) in enumerate(dataloader):

        # Compute prediction error
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        if batch % 100 == 0:
            loss, current = loss.item(), (batch + 1) * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")


def test(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()
    test_loss, correct = 0, 0
    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

### B6) Train until you reach at least 91% test accuracy

Tune:
- model size
- learning rate
- optimizer choice
- number of epochs
- batch size


```python
epochs = 5
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_dataloader, model, loss_fn, optimizer)
    test(test_dataloader, model, loss_fn)
print("Done!")

```
